```python
from scipy.special import polygamma

# Values of N to analyze
N_values = [8, 16, 32, 64, 128, 256]
# Upper bound error values for each N
upper_bound_errors = []

# Calculating the upper bound error for each N
for N in N_values:
    upper_bound_error = 4 * polygamma(1, N)
    upper_bound_errors.append(upper_bound_error)

# Print upper bound error values
list(zip(N_values, upper_bound_errors))
```
