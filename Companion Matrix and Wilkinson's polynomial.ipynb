```python
import numpy as np
from numpy.linalg import eig
from numpy.polynomial import Polynomial

# Given polynomial coefficients (assuming the leading coefficient is 1)
coefficients = [1]  # This represents x^n where n is the degree of the polynomial
coefficients.extend([-a for a in [1.2, -3.4, 5.6, -7.8]])  # Example coefficients for a_0, a_1, ..., a_{n-1}

# Construct the companion matrix
def construct_companion_matrix(coeffs):
    n = len(coeffs) - 1  # Degree of the polynomial
    companion_matrix = np.zeros((n, n))
    companion_matrix[:, -1] = -np.array(coeffs[:-1])  # Set last column to negative coefficients
    companion_matrix[1:, :-1] = np.eye(n - 1)  # Set sub-diagonal to 1's
    return companion_matrix

companion_matrix = construct_companion_matrix(coefficients)
print("Companion Matrix:")
print(companion_matrix)

# Verify that the characteristic polynomial is p(x)
p = Polynomial(coefficients)
characteristic_polynomial = np.poly(companion_matrix)
print("\nCharacteristic Polynomial Coefficients:")
print(characteristic_polynomial)
print("\nOriginal Polynomial Coefficients:")
print(p.coef)

# Check if the polynomials are the same (within a tolerance due to floating-point arithmetic)
print("\nAre the polynomials equivalent (within tolerance)?")
print(np.allclose(characteristic_polynomial, p.coef[::-1]))  # Reverse the coefficients of p for comparison
```

```python
# Wilkinson's polynomial coefficients can be computed as follows:
wilkinson_coeffs = np.poly(np.arange(1, 16))[::-1]

# Construct the companion matrix for Wilkinson's polynomial
companion_matrix_wilkinson = construct_companion_matrix(wilkinson_coeffs)

# Find eigenvalues of the companion matrix (which are the roots of the polynomial)
eigenvalues = eig(companion_matrix_wilkinson)[0]

# Find roots using the built-in method
roots = np.roots(wilkinson_coeffs)

# Calculate relative errors
def relative_error(true_values, approx_values):
    return np.abs((true_values - approx_values) / true_values)

relative_errors_eig = relative_error(np.arange(1, 16), np.sort(eigenvalues))
relative_errors_roots = relative_error(np.arange(1, 16), np.sort(roots))

print("Roots found using the companion matrix method:")
print(np.sort(eigenvalues))
print("\nRelative Error using the companion matrix method:")
print(relative_errors_eig)

print("\nRoots found using the built-in roots() method:")
print(np.sort(roots))
print("\nRelative Error using the built-in roots() method:")
print(relative_errors_roots)
```

```python
# Perturb Wilkinson's polynomial coefficients
wilkinson_coeffs_perturbed = wilkinson_coeffs.copy()
wilkinson_coeffs_perturbed[-2] -= 0.00001  # Change the coefficient of x^14

# Find roots of the perturbed polynomial
perturbed_roots = np.roots(wilkinson_coeffs_perturbed)

print("Roots of the perturbed Wilkinson's polynomial:")
print(perturbed_roots)

# Compare with unperturbed roots
print("\nChanges in roots due to perturbation:")
print(perturbed_roots - roots)
```
