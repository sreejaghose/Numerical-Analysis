```python
from scipy.integrate import simps, quadrature
import numpy as np

# Define the function for integration
def integrand(x):
    return np.exp(x)**2 * np.sin(x)**2



# Compute the integral using 3-point Gaussian quadrature
gaussian_approximation = quadrature(integrand, 0, 1, tol=1e-9, maxiter=3)[0]

gaussian_approximation

```

```python
# Function to count the number of matching digits between two numbers
def count_matching_digits(number1, number2):
    # Convert numbers to strings
    str1 = f"{number1:.10f}"
    str2 = f"{number2:.10f}"

    # Count matching digits
    matching_digits = 0
    for char1, char2 in zip(str1, str2):
        if char1 == char2:
            matching_digits += 1
        else:
            break
    
    return matching_digits

print(exact_integral_value)
print(gaussian_approximation)

# Count the matching digits between the exact integral value and the Gaussian quadrature approximation
matching_digits_gaussian = count_matching_digits(exact_integral_value, gaussian_approximation)

matching_digits_gaussian
print("2")

```
